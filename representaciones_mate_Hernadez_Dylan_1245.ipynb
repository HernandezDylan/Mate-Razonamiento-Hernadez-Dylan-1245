# Problema 1
La suma de un número más 9 es igual a 21Paso 1: Análisis y resolución algebraicaPara resolver este problema, seguimos una estructura lógica de despeje:Definir la variable: Sea $x$ el número desconocido.Plantear la ecuación: El enunciado dice que el número ($x$) sumado con 9 es igual a 21.$$x + 9 = 21$$Despejar la variable: Para dejar a $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad restando.$$x = 21 - 9$$Resultado:$$x = 12$$

```python
# Problema 1: Hallar el número desconocido
# total = 21
# constante = 9
resultado_total=int(input("Ingrese el resultado total: "))
numero_conocido=int(input("Ingrese el numero conocido: "))
# Despeje x = 21 - 9
x = resultado_total - numero_conocido

print(f"El número desconocido es: {x}")

```

#### Problema 2
Las edades de Pedro y su hermanaPaso 1: Análisis y resolución algebraicaEste problema requiere relacionar dos incógnitas basándonos en una sola variable:Definir las variables:Edad de la hermana: $x$Edad de Pedro: $2x$ (ya que es el doble que la de su hermana).Plantear la ecuación: La suma de ambas edades es 36.$$x + 2x = 36$$Simplificar: Sumamos los términos semejantes.$$3x = 36$$Despejar la variable: El 3 que multiplica a $x$ pasa dividiendo al 36.$$x = \frac{36}{3}$$$$x = 12$$Interpretación de resultados:Hermana ($x$): 12 años.Pedro ($2x$): $2 \cdot 12 = 24$ años.

```python
# Problema 2: Edades de Pedro y su hermana
# suma_total = 36
suma_total=int(input("Ingrese la suma total: "))
# La ecuación es: x + 2x = 36 -> 3x = 36
# Por lo tanto: x = suma_total / 3
edad_hermana = suma_total / 3
edad_pedro = edad_hermana * 2

print(f"La edad de la hermana es: {int(edad_hermana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
