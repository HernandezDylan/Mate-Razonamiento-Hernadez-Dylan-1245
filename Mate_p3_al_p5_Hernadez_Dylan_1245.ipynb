Problema 3
 Un número que se resta siete da como resultado 18Definir la variable: Sea $x$ el número desconocido.Plantear la ecuación: "Un número ($x$) menos ($-$) siete ($7$) es igual ($=$) a $18$".$$x - 7 = 18$$Despejar la variable: El 7 que está restando pasa al otro lado de la igualdad sumando.$$x = 18 + 7$$Resultado:$$x = 25$$

```python
# Problema 3: Hallar el número al que se le resta 7
resultado = 18
resta = 7

# Despeje: x = resultado + resta
# Convertimos a int para obtener un número entero exacto
numero = int(resultado + resta)

print(f"El número del problema 3 es: {numero}")
```

Problema 4
El triple de un número más cuatro es igual a 31Definir la variable: Sea $x$ el número buscado.Plantear la ecuación: "El triple de un número ($3x$) más ($+$) cuatro ($4$) es igual ($=$) a $31$".$$3x + 4 = 31$$Despejar la variable (paso A): Primero movemos el 4 que está sumando hacia el otro lado restando.$$3x = 31 - 4$$$$3x = 27$$Despejar la variable (paso B): El 3 que multiplica a $x$ pasa dividiendo al 27.$$x = \frac{27}{3}$$Resultado:$$x = 9$$

```python
# Problema 4: El triple de un número más 4 es igual a 31
total = 31
sumando = 4
multiplo = 3

# Despeje paso a paso:
# 1. Restar el sumando: 27
# 2. Dividir por el múltiplo: 27 / 3
# Usamos int() para transformar el resultado de la división (float) a entero
numero_final = int((total - sumando) / multiplo)

print(f"El número del problema 4 es: {numero_final}")
```

Problema 5
Dimensiones de un terreno rectangularEnunciado: Un terreno tiene un largo que mide cinco metros más que el ancho. Si el perímetro total es de 50 metros, ¿cuánto mide cada lado?Paso 1: Análisis y resolución algebraicaDefinir las variables:Sea $x$ el ancho del terreno.Entonces, el largo es $x + 5$ (ya que mide 5 metros más que el ancho).Plantear la ecuación:Recordemos que el perímetro de un rectángulo es la suma de sus cuatro lados ($2 \cdot \text{ancho} + 2 \cdot \text{largo}$).$$2(x) + 2(x + 5) = 50$$Simplificar la ecuación:Primero, aplicamos la propiedad distributiva (multiplicar el 2 por lo que está dentro del paréntesis):$$2x + 2x + 10 = 50$$Sumamos las $x$:$$4x + 10 = 50$$Despejar la variable (paso A):El 10 que está sumando pasa al otro lado restando:$$4x = 50 - 10$$$$4x = 40$$Despejar la variable (paso B):El 4 que multiplica a $x$ pasa dividiendo:$$x = \frac{40}{4}$$$$x = 10$$Interpretación de resultados:Ancho ($x$): 10 metros.Largo ($x + 5$): $10 + 5 = 15$ metros

```python
# Problema 5: Cálculo de dimensiones de un terreno
perimetro_total = 50
diferencia = 5

# Aplicamos la fórmula despejada: x = (Perímetro - 2*diferencia) / 4
# Envolvemos todo en int() para limpiar el resultado
ancho = int((perimetro_total - (2 * diferencia)) / 4)
largo = int(ancho + diferencia)

print(f"El ancho del terreno mide: {ancho} metros")
print(f"El largo del terreno mide: {largo} metros")
```
